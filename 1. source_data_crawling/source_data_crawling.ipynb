```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import time
import random
import os

# 300000000000~400000000000
# 600000000000~700000000000
# 파일당 2만개씩. 검색해서 근처 송장번호 기준으로 약 10개파일씩 크롤링
# 약 20만개 송장번호를 탐색
# 대략, 1.5만~2만개 원천데이터 수집 가능
file_num = 1000000
data_num = 100000
for i in range(3*file_num, 4*file_num):    
    # 입력할 파일 생성
    txt_name = str(i*data_num)+'_'+str(i*data_num+data_num-1)
    f_w = open('output_data/'+txt_name+'.txt', 'w', encoding = 'UTF-8')

    now = datetime.now()
    now_str = str(now.year) + str(now.month).zfill(2) + str(now.day)+str(now.hour).zfill(2)+str(now.minute).zfill(2)
    print(now_str, end=" : ")
    print(txt_name+" 수집 시작")
    time.sleep(2)
    f_w.write(now_str)
    f_w.write(" (yyyy.mm.dd.hh.mm)")
    f_w.write("\n")

    # 347176432174
    # 파일당 송장번호 1,000,000개 탐색
    for j in range(data_num):
        cj_add = 'https://www.doortodoor.co.kr/parcel/doortodoor.do?fsp_action=PARC_ACT_002&fsp_cmd=retrieveInvNoACT&invc_no='
        # 송장번호 = 12자리. = 100,000 * 1,000,000        
        invoice_num = i*data_num + j
        cj_add += str(invoice_num)        
        r = requests.get(cj_add)
        html = r.text
        soup = BeautifulSoup(html, 'html.parser')
        try :
            path = soup.find_all(title = "집배점 정보 팝업")
            len_path = len(path)
            date = soup.find_all("td")
            len_date = len(date)

            if j%2000==0:
                time.sleep(2)
                
            try:
                if date[-9].get_text()[73:77] == "배달완료":
                    pathlist = []
                    datelist = []
                    for k in range(0, len_path):
                        pathcell = path[k].get_text()
                        pathlist.append(pathcell)
                    # start date : date[6].get_text()
                    # end date : date[18].get_text()
                    try:
                        f_w.write(str(invoice_num))
                        f_w.write(", ")
                        f_w.write(pathlist[0])
                        f_w.write(", ")
                        f_w.write(pathlist[-1])
                        f_w.write("\n")
                        print(str(invoice_num)+" : 입력 완료")
                    except:
                        print(str(invoice_num)+" : 입력 중 에러")
            except:
                print(str(invoice_num)+" : 아직 배달완료되지 않았습니다.")
        except:
            print("집배점 정보 팝업이 없습니다.")
```
